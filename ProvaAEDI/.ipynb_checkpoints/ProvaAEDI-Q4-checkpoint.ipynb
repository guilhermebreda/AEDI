```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
# carregar base de dados
file_path = 'credit_customers.csv'
credit_data = pd.read_csv(file_path)

# verficar 5 primeiras linhas
credit_data.head()
```

## Regressão Logística

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# 2. Converter variáveis categóricas para numéricas usando LabelEncoder
label_encoder = LabelEncoder()
for col in credit_data.select_dtypes(include=['object']).columns:
    credit_data[col] = label_encoder.fit_transform(credit_data[col])

# 3. Separar variáveis independentes (X) e dependente (y)
X = credit_data.drop(columns=['class'])  # Variáveis preditoras
y = credit_data['class']  # Variável alvo (bom/mau pagador)

# 4. Normalizar os dados numéricos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. Dividir os dados em treino e teste (80% treino, 20% teste)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 6. Criar e treinar o modelo de Regressão Logística
model = LogisticRegression()
model.fit(X_train, y_train)

# 7. Fazer previsões no conjunto de teste
y_pred = model.predict(X_test)

# Calcular métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Exibir métricas
print(f"Acurácia: {accuracy:.2f}")
print(f"Precisão: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")

# 8. Exibir relatório de classificação
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred))

# 9. Matriz de Confusão
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues", xticklabels=['Bom Pagador', 'Mau Pagador'], yticklabels=['Bom Pagador', 'Mau Pagador'])
plt.xlabel("Previsto")
plt.ylabel("Real")
plt.title("Matriz de Confusão")
plt.show()
```

## Árvore de decisão

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Criar e treinar o modelo de Árvore de Decisão
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred_tree = tree_model.predict(X_test)

# Avaliação do modelo
accuracy_tree = accuracy_score(y_test, y_pred_tree)
precision_tree = precision_score(y_test, y_pred_tree)
recall_tree = recall_score(y_test, y_pred_tree)
f1_tree = f1_score(y_test, y_pred_tree)

# Exibir métricas
print(f"Acurácia: {accuracy_tree:.2f}")
print(f"Precisão: {precision_tree:.2f}")
print(f"Recall: {recall_tree:.2f}")
print(f"F1-Score: {f1_tree:.2f}")

# Relatório completo
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred_tree))

# Matriz de Confusão
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred_tree), annot=True, fmt="d", cmap="Blues",
            xticklabels=['Bom Pagador', 'Mau Pagador'], yticklabels=['Bom Pagador', 'Mau Pagador'])
plt.xlabel("Previsto")
plt.ylabel("Real")
plt.title("Matriz de Confusão - Árvore de Decisão")
plt.show()
```

## Random Forest Trees

```python
from sklearn.ensemble import RandomForestClassifier

# Criar e treinar o modelo de Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred_rf = rf_model.predict(X_test)

# Avaliação do modelo
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)

# Exibir métricas
print(f"Acurácia: {accuracy_rf:.2f}")
print(f"Precisão: {precision_rf:.2f}")
print(f"Recall: {recall_rf:.2f}")
print(f"F1-Score: {f1_rf:.2f}")

# Relatório completo
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred_rf))

# Matriz de Confusão
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt="d", cmap="Blues",
            xticklabels=['Bom Pagador', 'Mau Pagador'], yticklabels=['Bom Pagador', 'Mau Pagador'])
plt.xlabel("Previsto")
plt.ylabel("Real")
plt.title("Matriz de Confusão - Random Forest")
plt.show()
```

```python
import shap

# Criar o objeto explicador para o modelo Random Forest
explainer = shap.Explainer(rf_model, X_train)

# Calcular os valores SHAP desativando a verificação de aditividade
shap_values = explainer(X_test, check_additivity=False)

# Resumo gráfico das variáveis mais importantes
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X_test, plot_type="bar")
```

```python
import shap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criar o objeto explicador para o modelo Random Forest
explainer = shap.TreeExplainer(rf_model)

# Calcular os valores SHAP
shap_values = explainer.shap_values(X_test_df)

# Verificar se shap_values é uma lista (indica múltiplas classes)
if isinstance(shap_values, list):
    shap_values_selected = shap_values[1]  # Pegamos a classe 1 (inadimplente)
else:
    shap_values_selected = shap_values  # Para modelos binários que retornam um único array

# Garantir que shap_values_selected tenha a mesma quantidade de colunas que X_test_df
if shap_values_selected.ndim == 3:  
    shap_values_selected = shap_values_selected[:, :, 0]  # Remover dimensões extras, se existirem

# Calcular a média absoluta dos valores SHAP
shap_mean = np.abs(shap_values_selected).mean(axis=0)

# Garantir que shap_mean seja um array 1D
shap_mean = np.squeeze(shap_mean)

# Criar DataFrame com os nomes das features e a média SHAP
if len(shap_mean) == len(X_test_df.columns):
    shap_df = pd.DataFrame({'Feature': X_test_df.columns, 'SHAP Mean': shap_mean})

    # Ordenar por importância
    shap_df = shap_df.sort_values(by='SHAP Mean', ascending=False)

    # Exibir DataFrame ordenado
    print(shap_df)

    # Criar gráfico de barras
    plt.figure(figsize=(10, 6))
    sns.barplot(x='SHAP Mean', y='Feature', data=shap_df, palette='coolwarm')
    plt.title("Importância das Variáveis Segundo SHAP")
    plt.show()
else:
    print(f"Erro: O número de colunas em shap_values ({len(shap_mean)}) não corresponde ao de X_test_df ({len(X_test_df.columns)}).")
```
