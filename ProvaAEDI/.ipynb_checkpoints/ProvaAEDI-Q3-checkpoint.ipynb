## Filtrar colunas relevantes p/ ANOVA

```python
# Selecionar as colunas relevantes, incluindo "Country" para ANOVA
selected_cols = ["Quantity", "Price", "Country"]
filtered_data = retail_data[selected_cols]
```

```python
# Filtrar valores inconsistentes (excluir quantidades e pre√ßos negativos ou zero)
filtered_data = filtered_data[(filtered_data["Quantity"] > 0) & (filtered_data["Price"] > 0)]
```

## ANOVA

```python
# 1. Remove missing values and filter relevant columns
anova_data = filtered_data[['Country', 'Quantity', 'Price']].dropna()

# 2. Perform ANOVA for Quantity
quantity_groups = [group["Quantity"].values for _, group in anova_data.groupby("Country")]
f_stat_quantity, p_value_quantity = stats.f_oneway(*quantity_groups)

# 3. Perform ANOVA for Price
price_groups = [group["Price"].values for _, group in anova_data.groupby("Country")]
f_stat_price, p_value_price = stats.f_oneway(*price_groups)

# 4. Display Results
print("\nüîπ **ANOVA for Quantity Sold**")
print(f"F-Statistic: {f_stat_quantity:.2f}, P-Value: {p_value_quantity:.4f}")

print("\nüîπ **ANOVA for Product Price**")
print(f"F-Statistic: {f_stat_price:.2f}, P-Value: {p_value_price:.4f}")

# 5. Interpretation
if p_value_quantity < 0.05:
    print("\n‚úÖ Quantity differences between countries are statistically significant!")
else:
    print("\n‚ùå No significant differences found in quantity sold among countries.")

if p_value_price < 0.05:
    print("\n‚úÖ Price differences between countries are statistically significant!")
else:
    print("\n‚ùå No significant differences found in product prices among countries.")
```

### Teste Shapiro-Wilk e Levene

```python
import scipy.stats as stats

# 1. Check Normality of Residuals using Shapiro-Wilk Test
anova_data["Quantity_Residuals"] = anova_data["Quantity"] - anova_data.groupby("Country")["Quantity"].transform("mean")
anova_data["Price_Residuals"] = anova_data["Price"] - anova_data.groupby("Country")["Price"].transform("mean")

shapiro_quantity = stats.shapiro(anova_data["Quantity_Residuals"])
shapiro_price = stats.shapiro(anova_data["Price_Residuals"])

print("\nüîπ **Shapiro-Wilk Test for Normality**")
print(f"Quantity Residuals - W: {shapiro_quantity[0]:.4f}, P-Value: {shapiro_quantity[1]:.4f}")
print(f"Price Residuals - W: {shapiro_price[0]:.4f}, P-Value: {shapiro_price[1]:.4f}")

# 2. Check Homogeneity of Variances using Levene‚Äôs Test
levene_quantity = stats.levene(*quantity_groups)
levene_price = stats.levene(*price_groups)

print("\nüîπ **Levene‚Äôs Test for Homogeneity of Variances**")
print(f"Quantity - W: {levene_quantity[0]:.2f}, P-Value: {levene_quantity[1]:.4f}")
print(f"Price - W: {levene_price[0]:.2f}, P-Value: {levene_price[1]:.4f}")

# ‚úÖ 3. Interpretation
if shapiro_quantity[1] < 0.05:
    print("\n‚ùå Quantity residuals are NOT normally distributed!")
else:
    print("\n‚úÖ Quantity residuals follow a normal distribution.")

if shapiro_price[1] < 0.05:
    print("‚ùå Price residuals are NOT normally distributed!")
else:
    print("‚úÖ Price residuals follow a normal distribution.")

if levene_quantity[1] < 0.05:
    print("\n‚ùå Variance in quantity is NOT homogeneous across countries!")
else:
    print("\n‚úÖ Variance in quantity is homogeneous across countries.")

if levene_price[1] < 0.05:
    print("‚ùå Variance in price is NOT homogeneous across countries!")
else:
    print("‚úÖ Variance in price is homogeneous across countries.")
```

### Alternativa a ANOVA

```python
# 1. Kruskal-Wallis Test (for non-normal data)
kw_quantity = stats.kruskal(*quantity_groups)
kw_price = stats.kruskal(*price_groups)

print("\nüîπ **Kruskal-Wallis Test for Non-Normal Data**")
print(f"Quantity - H: {kw_quantity[0]:.2f}, P-Value: {kw_quantity[1]:.4f}")
print(f"Price - H: {kw_price[0]:.2f}, P-Value: {kw_price[1]:.4f}")

# 2. Welch‚Äôs ANOVA (for unequal variances)
welch_quantity = stats.ttest_ind(*quantity_groups, equal_var=False)
welch_price = stats.ttest_ind(*price_groups, equal_var=False)

print("\nüîπ **Welch‚Äôs ANOVA for Unequal Variances**")
print(f"Quantity - T: {welch_quantity[0]:.2f}, P-Value: {welch_quantity[1]:.4f}")
print(f"Price - T: {welch_price[0]:.2f}, P-Value: {welch_price[1]:.4f}")
```
